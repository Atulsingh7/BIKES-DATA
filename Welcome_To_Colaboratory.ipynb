```
salesTable2['R_rank'] = salesTable2['Recency'].rank(ascending=False)
salesTable2['F_rank'] = salesTable2['Frequency'].rank(ascending=True)
salesTable2['M_rank'] = salesTable2['Monetary'].rank(ascending=True)
 
# normalizing the rank of the customers
salesTable2['R_rank_norm'] = (salesTable2['R_rank']/salesTable2['R_rank'].max())*100
salesTable2['F_rank_norm'] = (salesTable2['F_rank']/salesTable2['F_rank'].max())*100
salesTable2['M_rank_norm'] = (salesTable2['F_rank']/salesTable2['M_rank'].max())*100
 
#rfm.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
 
salesTable2.head(10)
```

```
salesTable2[salesTable2['Frequency'] > 1].count()
```

```
rfm = salesTable2 
```

```
rfm['RFM_Score'] = 0.15*rfm['R_rank_norm']+0.28 * rfm['F_rank_norm']+0.57*rfm['M_rank_norm']
rfm['RFM_Score']*=0.05
rfm = rfm.round(2)
rfm.head(7)
```

```
cnt = 0
for i in rfm['RFM_Score']:
  if i<4 and i>1:
    cnt+=1
cnt
```

```
rfm["Customer_segment"] = np.where(rfm['RFM_Score'] >
									4.5, "Top Customers",
									(np.where(
										rfm['RFM_Score'] > 3.5,
										"High value Customer",
										(np.where(
	rfm['RFM_Score'] > 2.5,
							"Medium Value Customer",
							np.where(rfm['RFM_Score'] > 2.0,
							'Low Value Customers', 'Lost Customers'))))))
rfm.head(2)
```
